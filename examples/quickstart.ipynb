# pptk-revived Quickstart

This notebook demonstrates basic usage of **pptk-revived** for point cloud visualization.

> **Note:** The viewer opens in a separate window. In headless environments set `DISPLAY` or use `Xvfb`.

## Imports

You can import the package as either `pptk` or `pptk_revived` — both are equivalent.

```python
import numpy as np
import pptk

print('pptk version:', pptk.__version__)
```

```python
import pptk_revived

# pptk_revived re-exports everything from pptk — use whichever name you prefer
print('pptk_revived version:', pptk_revived.__version__)
print('viewer:', pptk_revived.viewer)
print('estimate_normals:', pptk_revived.estimate_normals)
```

## Create a random point cloud

```python
rng = np.random.default_rng(42)
xyz = rng.random((10000, 3)).astype(np.float32)
print(f'Point cloud shape: {xyz.shape}')
```

## Visualize with pptk.viewer()

```python
v = pptk.viewer(xyz)
v.set(point_size=0.005)
```

## Color points by a scalar attribute (height)

```python
v2 = pptk.viewer(xyz, xyz[:, 2])  # shade by Z (height)
v2.set(point_size=0.005)
v2.color_map('jet')
```

## Color points with RGB values

```python
rgb = rng.random((len(xyz), 3)).astype(np.float32)
v3 = pptk.viewer(xyz, rgb)
v3.set(point_size=0.005)
```

## Close viewers

```python
v.close()
v2.close()
v3.close()
```
